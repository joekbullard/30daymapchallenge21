```python
import requests
import json
import contextily as cx
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
overpass_url = "http://overpass-api.de/api/interpreter"
overpass_q = """
[out:json][timeout:900];
area[name="France"]->.searchArea;
(
  node["landuse"="vineyard"](area.searchArea)(42.065607,-7.163086,51.371780,9.184570);
  way["landuse"="vineyard"](area.searchArea)(42.065607,-7.163086,51.371780,9.184570);
  relation["landuse"="vineyard"](area.searchArea)(42.065607,-7.163086,51.371780,9.184570);
);
out body;
>;
out center;
"""

response = requests.get(overpass_url, 
                        params={'data': overpass_q})

data = response.json()
```

```python
coords = []
for element in data['elements']:
    if element['type'] == 'node':
        lon = element['lon']
        lat = element['lat']
        coords.append(Point(lon, lat))
    elif 'center' in element:
        lon = element['center']['lon']
        lat = element['center']['lat']
        coords.append(Point(lon, lat))
        
```

```python
coords = []
for element in data['elements']:
    if element['type'] == 'node':
        lon = element['lon']
        lat = element['lat']
        coords.append(Point(lon, lat))
    elif 'center' in element:
        lon = element['center']['lon']
        lat = element['center']['lat']
        coords.append(Point(lon, lat))
        
df = pd.DataFrame(coords)
df = df.rename(columns={0:'geom'})
gdf = gpd.GeoDataFrame(df, crs=4326, geometry='geom')
#gdf = gdf.to_crs(27700)
gdf.crs
```
